```python
tweets = []
n_tweets=2000

for i, tweet in tqdm(enumerate(scraper.get_items()), total=n_tweets):
  data = [tweet.date, tweet.id, tweet.content]
  tweets.append(data)
  if i>n_tweets:
    break
```

```python
tweet_df = pd.DataFrame(tweets, columns=["date","id","content"])

# tweet_df.to_csv("congress_tweets.csv",index=False)
```

```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
import numpy as np
from scipy.special import softmax
```

```python
# Preprocess text (username and link placeholders)
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
# PT
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
#model.save_pretrained(MODEL)
text = "They are the Inspiration Strength Motivation to workers &amp; party with guidance of Soniaji-Rahul We have to fight in same spirit at ZP-TP BBMP Co-Op Societies &amp; MP elections Congress is the hope to save India and Democracy #Congress #aicc #media #LoksabhaElection2024 #Awareness https://t.co/QZ41JMpyes"
text = preprocess(text)
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
scores = output[0][0].detach().numpy()
scores = softmax(scores)
```

```python
ranking = np.argsort(scores)
ranking = ranking[::-1]
for i in range(scores.shape[0]):
    l = config.id2label[ranking[i]]
    s = scores[ranking[i]]
    print(f"{i+1}) {l} {np.round(float(s), 4)}")
```

```python
import pandas as pd
import torch

# List of texts
texts = tweet_df['content'].tolist()

# Create an empty dataframe
labeledf = pd.DataFrame(columns=["Text", "Sentiment"])


# Iterate over the texts
for text in texts:
    # Preprocess text
    preprocessed_text = preprocess(text)
  
    # Tokenize and encode the preprocessed text
    encoded_input = tokenizer(preprocessed_text, truncation=True, padding=True, max_length=512, return_tensors='pt')
    
    # Get the model's output
    with torch.no_grad():
        output = model(**encoded_input)
    scores = output.logits[0].detach().numpy()
    scores = softmax(scores)
  
    # Get the highest ranked sentiment label
    highest_rank = np.argmax(scores)
    sentiment_label = config.id2label[highest_rank]
    
    # Save the text and sentiment label in the dataframe
    labeledf = labeledf.append({"Text": text, "Sentiment": sentiment_label}, ignore_index=True)

# Print the dataframe
print(labeledf)
```
